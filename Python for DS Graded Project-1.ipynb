# Data Cleaning and Preprocessing

##### 4.Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix

```python
df['Value'] = df['Value'].str.extract('€(.*)')
df['Wage'] = df['Wage'].str.extract('€(.*)')
df['Release Clause'] = df['Release Clause'].str.extract('€(.*)')
```

```python
df['Value'] = (df.Value.replace(r'[KM]+$', '', regex=True).astype(float) * \
                df.Value.str.extract(r'[\d\.]+([KM]+)', expand=False)
                .fillna(1)
                .replace(['K','M'], [10**3, 10**6]).astype(int))

df['Wage'] = (df.Wage.replace(r'[KM]+$', '', regex=True).astype(float) * \
                df.Wage.str.extract(r'[\d\.]+([KM]+)', expand=False)
                .fillna(1)
                .replace(['K','M'], [10**3, 10**6]).astype(int))

df['Release Clause'] = (df['Release Clause'].replace(r'[KM]+$', '', regex=True).astype(float) * \
                df['Release Clause'].str.extract(r'[\d\.]+([KM]+)', expand=False)
                .fillna(1)
                .replace(['K','M'], [10**3, 10**6]).astype(int))
```

###### 5.Convert the column "Joined" into integer data type with keeping only the year

```python
df['Joined']=df['Joined'].str.extract('.*, (.*)').astype(float)
```

##### 6.Convert the column "Contract Valid Until" to pandas datetime type

```python
df['Contract Valid Until']= pd.to_datetime(df['Contract Valid Until'])
df['Contract Valid Until']
```

##### 7.The column 'Height' is in inches with a quotation mark, Convert to float with decimal points

```python
df['Height']=df['Height'].replace({'\'': '.'}, regex=True).astype(float)
df['Height']
```

##### 8.The column "Weight" has the suffix as lbs, remove the suffix and convert to float. (2 points)

```python
df['Weight']=df['Weight'].str.extract('(.*)lbs').astype(float)
df['Weight']
```
